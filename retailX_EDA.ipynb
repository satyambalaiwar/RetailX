```python
print("Total Revenue:", df['total_sale'].sum())
print("Total Profit:", df['profit'].sum())
print("Average Profit per Transaction:", df['profit'].mean())
```

```python
df.groupby('product_category')['total_sale'].sum().sort_values(ascending=False).plot(kind='bar', figsize=(10, 4), title='Revenue by Product Category')
plt.ylabel("Total Sale")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x='age_group', y='profit', estimator=sum, ci=None)
plt.title('Profit Contribution by Age Group')
plt.ylabel("Total Profit")
plt.tight_layout()
plt.show()
```

```python
df.groupby('month')['total_sale'].sum().plot(marker='o', title='Monthly Sales Trend')
plt.ylabel("Total Sale")
plt.show()

df.groupby('day_of_week')['total_sale'].sum().plot(kind='bar', title='Sales by Day of the Week')
plt.ylabel("Total Sale")
plt.show()

df.groupby('hour')['total_sale'].sum().plot(kind='line', title='Hourly Sales Trend')
plt.ylabel("Total Sale")
plt.xticks(range(0, 24))
plt.show()
```

```python
sns.boxplot(data=df, x='gender', y='total_sale')
plt.title("Sales Distribution by Gender")
plt.show()
```
